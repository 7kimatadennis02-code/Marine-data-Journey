## Simple if statement

```python
water_temp = 32.5

if water_temp > 30:
    print("Warning: Temperature too high")
    print("Coral bleaching risk detected")
```

## if and else

```python
fish_count = 35
if fish_count >= 50:
    print("Fish population: Healthy")
else:
    print("Fish population: Below threshold")
    print("Conservation action may be required")
```

## if, elif and else

```python
ph_level = 7.8

if ph_level < 6.5:
    print("pH too low - acidic water")
    print("Marine life at risk")
elif ph_level > 8.5:
    print("pH too high - alkaaline water")
    print("Check for pollution source")
else:
    print("pH level normal")
    print("Water chemistry within safe range")
```

# Nested Conditionals

```python
water_temp = 29.0
salinity = 38.5

if water_temp > 28:
    print("Temperature elevated")
    if salinity > 36:
        print("High salinity detected")
        print("Combined stress - urgent monitoring needed")
    else:
        print("Salinity normal")
        print("Monitor temperature closely")
else:
    print("All parameters within normal range")
```

# Multiple conditions - and / or

```python
dissolved_oxygen = 4.2
turbidity = 18.5

# and - both conditions must be True
if dissolved_oxygen < 5.0 and turbidity > 15:
    print("Poor water quality detected")
    print("Low oxygen AND high turbidity")

# or - at least one condition must be True
if dissolved_oxygen < 5.0 or turbidity > 15:
    print("At least one parameter out of range")
    print("Investgation recommended")
```
