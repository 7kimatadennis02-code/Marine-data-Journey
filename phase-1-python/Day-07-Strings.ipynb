## String indexing and slicing

```python
site = "Watamu Marine Reserve"

# Indexing - access one character
print(f"First character: {site[0]}")
print(f"Last character: {site[-1]}")
print(f"Fifth character: {site[4]}")

# Slicing - access a range
print(f"\nFirst 6 characters: {site[0:6]}")
print(f"Characters 7 to 12: {site[7:13]}")
print(f"Last 7 characters: {site[-7:]}")
print(f"Every other character: {site[::2]}")

# String length
print(f"\nLength: {len(site)} characters")

# Check start and end
print(f"\nStarts with 'Watamu': {site.startswith('Watamu')}")
print(f"Ends with 'Reserve': {site.endswith('Reserve')}")
```

## Cleaning Text

```python
# Messy data froma field sheet
messy_site = " watamu marine reserve "
messy_species = "GREEN TURTLE"
messy_code = "WTM-001 "
inconsistent = "mombasa Marine RESERVE"

# Strip whitespace
clean_site = messy_site.strip()
print(f"Stripped: '{clean_site}'")

# Standardize capitalisation
print(f"Lower: {messy_species.lower()}")
print(f"Upper: {messy_species.upper()}")
print(f"Title: {messy_species.title()}")
print(f"Capitalize: {messy_species.capitalize()}")

# Fix inconsistent capitalisation
print(f"\nFixed: {inconsistent.title()}")

# Strip and clean code
print(f"Clean code: '{messy_code.strip()}")

# Replace characters
messy_date = "2026/04/07"
clean_date = messy_date.replace("/", "-")
print(f"\nDate fixed: {clean_date}")

messy_value = "27.5 °C"
clean_value = messy_value.replace(" ", "")
print(f"Value fixed: {clean_value}")
```

## Splitting and Joining

```python
# Split a string into a list
site_string = "Watamu,Malindi,Kilifi,Mombasa,Lamu"
sites_list = site_string.split(",")
print(f"Split: {sites_list}")
print(f"Count: {len(sites_list)}")

# Access individual items after split
print(f"First: {sites_list[0]}")
print(f"Last: {sites_list[-1]}")

# Split on spaces
species = "Green Turtle Hawksbill Turtle Dugong"
words = species.split(" ")
print(f"\nWords: {words}")

# Join a list back into a string
joined = " | ".join(sites_list)
print(f"|nJoined with |: {joined}")

comma_joined = ", ".join(sites_list)
print(f"Joined with comma: {comma_joined}")

# Split a data row
data_row = "Watamu|7.8|27.5|87|Normal"
fields = data_row.split("|")
print(f"\nData row fields:")
for i, field in enumerate(fields):
    print(f" Field {i}: {field}")
```

## Finding and Checking

```python
notes = "Water temperature elevated at Watamu site. \
Coral bleachng observed near reef. \
pH readng normal. Follow up required."

# Check substring exists
print(f"Contains 'bleaching': {'bleaching' in notes}")
print(f"Contains 'shark': {'shark' in notes}")

# Find position of a word
position = notes.find("Watamu")
print(f"\n'Watamu' found at position: {position}")

# Count occurences
count = notes.count("normal")
print(f"'normal' appears: {count} time(s)")

# Check string content
code = "WTM001"
number = "12345"
print(f"\n'{code}' is alphanumeric: {code.isalnum()}")
print(f"'{number}' is numeric: {number.isnumeric()}")

# Count words
word_count = len(notes.split())
print(f"\nWord count in notes; {word_count}")
```

## String Formatting

```python
# f-strings with formatting options
site = "Watamu"
ph = 7.823456
temp = 27.5
fish_count = 87
status = "Normal"

# Width and alignment
print(f"{'Site':<15} {'pH':>8} {'Temp':>8} {'Fish':>6} {'Status':<10}")
print("=" * 50)
print(f"{site:<15} {ph:>8.2f} {temp:>8.1f} {fish_count:>6} {status:<10}")

# Building a report string
report = (f"Site Report for {site}\n"
          f"pH: {ph:.2f} | "
          f"Temperature: {temp}°C | "
          f"Fish count: {fish_count}")
print(f"\n{report}")

# Padding numbers with zeros
station_id = 7
formatted_id = f"STN{station_id:03d}"
print(f"\nStation ID: {formatted_id}")
```
