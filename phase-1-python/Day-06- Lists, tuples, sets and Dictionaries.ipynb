## Lists

```python
# Creating a List
monitoring_sites = ["Watamu", "Malindi", "Kilifi",
                    "Mombasa", "Lamu"]

#Accessing items by index _starts at 0
print(f"First site: {monitoring_sites[0]}")
print(f"Last site: {monitoring_sites[-1]}")
print(f"Second to fourth: {monitoring_sites[1:4]}")

# List length
print(f"Total sites: {len(monitoring_sites)}")

# Adding items
monitoring_sites.append("Shimba")
print(f"After adding: {monitoring_sites}")

# Removing items
monitoring_sites.remove("Kilifi")
print(f"After removing: {monitoring_sites}")

## Changing an item
monitoring_sites[0] = "Watamu Marine Reserve"
print(f"After updating: {monitoring_sites}")

# Check if item exists
print(f"Malindi in list: {'Malindi' in monitoring_sites}")
```

## Tuples

```python
# Coordinates should never change - use a tuple
watamu_coords = (-3.3053, 40.0169)
malindi_coords = (-3.2138, 401169)
mombasa_coords = (-4.0435, 39.6682)

print(f"Watamu coordinates: {watamu_coords}")
print(f"Latitude: {watamu_coords[0]}")
print(f"Longitude: {watamu_coords[1]}")

# Tuple of fixed reference values
safe_ranges = (6.5, 8.5, 5.0, 30.0)
#ph_min, ph_max, oxygen_min, temp_max

print(f"\nSafe pH range: {safe_ranges[0]} to {safe_ranges[1]}")
print(f"Min oxygen: {safe_ranges[2]} mg/l")
print(f"Max temperature: {safe_ranges[3]}°C")

# Tuple unpacking  - very useful
ph_min, ph_max, oxygen_min, temp_max = safe_ranges
print(f"\nUnpacked pH min: {ph_min}")
print(f"Unpacked pH max: {ph_max}")
```

## Sets

```python
# Speces recorded accross multiple surveys
survey_1 = {"Green Turtle", "Dugong", "Dolphin",
            "Manta Ray", "Whale Shark"}

survey_2 = {"Green Turtle", "Hawksbill Turtle",
            "Dolphin", "Humpback Whale", "Dugong"}

# Sets remove duplicates automatically
species_records = ["Green Turtle", "Dolphin",
                   "Green Turtle", "Dugong",
                   "Dolphin", "Manta Ray",
                   "Green Turtle"]

unique_species = set(species_records)
print(f"All recirds: {species_records}")
print(f"Unique species: {unique_species}")
print(f"Total records: {len(species_records)}")
print(f"unique count: {len(unique_species)}")

# Set operations _very useful for surveys
print(f"\nSpecies in both surveys:")
print(survey_1 & survey_2)      # intersection

print(f"\nAll species across both surveys:")
print(survey_1 | survey_2)     # union

print(f"\nOnly in survey1:")
print(survey_1 - survey_2)     # difference

# Membership check - very fast
print(f"\nDugong recorded: {'Dugong' in survey_1}")
print(f"Orca recorded: {'Orca' in survey_1}")
```

## Dictionaries

```python
# site information stored as a dictionary
watamu = {
    "name": "Watamu Marine Reserve",
    "latitude": -3.3053,
    "longitude": 40.0169,
    "area_km2": 10,
    "established": 1968,
    "ph": 7.8,
    "water_temp": 27.5,
    "fish_species": 87
}


# Accessing  values by key
print(f"Site: {watamu['name']}")
print(f"Established: {watamu['established']}")
print(f"Fish species: {watamu['fish_species']}")

# Adding a new key
watamu["coral_cover_percent"] =42
print(f"\nCoral cover added: {watamu['coral_cover_percent']}%")

# Updating an existing value
watamu["ph"] = 7.9
print(f"Updated pH: {watamu['ph']}")

# Check if key exists
print(f"\n'water_temp' exists: {'water_temp' in watamu}")
print(f"'salinity' exists: {'salinity' in watamu}")

# Loop through a dictionary
print("\nFull site profile:")
for key, value in watamu.items():
    print(f" {key}: {value}")
```

## List of Dictionaries - Data Table Pattern

```python
# Multiple sites stored as a list of dictionaries
sites = [
    {"name": "Watamu", "ph": 7.8,
     "temp": 27.5, "fish_count": 87},
    {"name": "Malindi", "ph": 6.3,
     "temp": 31.2, "fish_count":45},
    {"name": "Kilifi", "ph": 7.5,
     "temp": 26.8, "fish_count":112},
    {"name": "Mombasa", "ph": 8.1,
     "temp": 29.4, "fish_count": 78},
    {"name": "Lamu", "ph": 8.7,
     "temp": 28.1, "fish_count":63},
]

print("Site Summary Report")
print("=" * 45)

for site in sites:
    name = site["name"]
    ph = site["ph"]
    temp = site["temp"]
    fish = site["fish_count"]

    # Flag any issues
    alerts = []
    if ph < 6.5 or ph > 8.5:
        alerts.append("pH out of range")
    if temp > 30:
        alerts.append("temp too high")
    if fish < 50:
        alerts.append("low fish count")

    status = "⚠️ ALERT" if alerts else "✅ Normal"
    print(f"{name:<10} pH:{ph} "
          f"Temp:{temp}°C Fish:{fish} - {status}")
    for alert in alerts:
        print(f"           › {alert}")
```
