## Exercise 2 - Fising Zone Alert System

```python
# Fishing zone data
zone = "Malindi Fishing Zone"
daily_catch_kg = 280
vessel_count = 12
protected_area_distance_km = 2.3
fishing_license = True

print(f"Zone: {zone}")
print("=" * 35)

# Catch threshold
if daily_catch_kg > 500:
    catch_alert = "Overfishing risk - reduce effort"
elif daily_catch_kg > 300:
    catch_alert = "Approachingh limit - monitor closely"
elif daily_catch_kg > 100:
    catch_alert = "Within sustainable range"
else:
    catch_alert = "Low catch - investigate cause"
print(f"Daily catch: {daily_catch_kg} kg - {catch_alert}")

# Vessel count
if vessel_count > 20:
    vessel_alert = "High vessel density - crowding risk"
elif vessel_count > 10:
    vessel_alert = "Moderate vessel activity"
else:
    vessel_alert = "Low vessel activity"
print(f"Vessels: {vessel_count} vessels - {vessel_alert}")

# Protected area proximity
if protected_area_distance_km < 1.0:
    proximity_alert = "Violation - inside protected zone"
elif protected_area_distance_km < 2.0:
    proximity_alert = "Warning - very close to protected area"
else:
    proximity_alert = "Safe distance from protected area"
print(f"Disatance to protected area: {protected_area_distance_km}km - {proximity_alert}")

# License check
if fishing_license:
    print("Fishing license: Valid")
else:
    print("Fishing license: INVALID - vassel must leave zone")
```

## Exercise 4 - Overall Site Conservation Score

```python
# Combine multiple factors into one overall score
site = "Watamu Marine Reserve"

# Individual scires out of 10
water_quality_score = 7
biodiversity_score = 8
human_impact_score = 5
habitat_condition_score = 6

# Calculate overall average
overall_score = (water_quality_score + biodiversity_score + human_impact_score + 
                 habitat_condition_score) / 4
print(f"Site: {site}")
print("=" * 35)
print(f"Water quality: {water_quality_score}/10")
print(f"Biodiversity: {biodiversity_score}/10")
print(f"Human impact: {human_impact_score}/10")
print(f"Overall score: {overall_score:.1f}/10")
print("---")

# Overall classification
if overall_score >= 8:
    print("Conservation status: EXCELLENT - maintain currennt protection")
elif overall_score >= 6:
    print("Conservation status: GOOD - minor improvements needed")
elif overall_score >= 4:
    print("Conservation status: FAIR - active management required")
else:
    print("Conservation status: POOR - urgent intervention needed")
```
