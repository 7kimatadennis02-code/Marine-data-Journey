## Exercise 1 - Field Data Cleaner

```python
# Messy field data entries taht need cleaning
raw_entries = [
    "  watamu marine reserve  ",
    "MALINDI REEF STATION",
    "kilifi channel",
    "  Mombasa Harbour  ",
    "lamu archipelago BEACH",
    "SHIMBA hills marine",
]

print("Field Data Cleaner")
print("=" * 40)
print(f"{'Original':<30} {'Cleaned':<30}")
print("-" * 40)

cleaned_entries = []

for entry in raw_entries:
    cleaned = entry.strip().title()
    cleaned_entries.append(cleaned)
    print(f"{entry.strip():<30} {cleaned:<30}")
    
print(f"\nTotal entries cleaned: {len(cleaned_entries)}")
print(f"Cleaned list: {cleaned_entries}")
```

## Exercise 2 - Species Name Standardizer

```python
# Inconsistent species name from different data sources
raw_species = [
    "green turtle",
    "Green Turtle",
    "Green turtle",
    "hawksbill turtle",
    "HAWKSBILL TURTLE",
    "Hawksbill Turtle",
    "dugong",
    "DUGONG",
    "whale shark",
    "Whale Shark",
    "WHALE SHARK",
]

print("Species Name Standardizer")
print("=" * 40)

# Standardize and find unique names
standardized = []
for name in raw_species:
    standard = name.strip().title()
    standardized.append(standard)

unique_standard = set(standardized)

print(f"Original entries: {len(raw_species)}")
print(f"After standardizing: {len(unique_standard)} unique species")
print(f"\nUnique species:")
for species in sorted(unique_standard):
    print(f" {species}")
```

## Data Row Paser

```python
# Raw data rows from a csv_like file
raw_data = [
    "watamu|2026-04-04|7.8|27.5|6.2|Normal",
    "Malindi|2026-04-01|6.3|31.2|4.8|Alert",
    "Kilifi|2026-04-02|7.5|26.8|7.1|Normal",
    "Mombasa|2026-04-02|8.1|29.4|5.5|Normal",
    "Lamu|2026-04-03|8.7|28.1|4.2|Alert",
]

print("Parsed Water Quality Data")
print("=" * 55)
print(f"{'Site':<10} {'Date':<12} "
      f"{'pH':>5} {'Temp':>6} {'02':>5} {'Status':<8}")
print("-" * 55)

alert_sites = []

for row in raw_data:
    fields = row.split("|")
    site = fields[0]
    date = fields[1]
    ph = float(fields[2])
    temp = float(fields[3])
    oxygen = float(fields[4])
    status = fields[5]

    print(f"{site:<10} {date:<12} "
          f"{ph:>5.1f} {temp:>6.1f} {oxygen:>5.1f} {status:<8}")

    if status == "Alert":
        alert_sites.append(site)

print("-" * 55)
print(f"Alert sites: {', '.join(alert_sites)}")
```

## Exercise 4 - Field notes analyzer

```python
# Field notes from a survey
field_notes = """
Survey date: 2026-04-07
Location: Watamu Marine Reserve
Observer: Dennis Kimata
Conditions: Calm sea, good vsibility
Observations: Green turtle spotted near reef.
Water temperature elevated above normal range.
Coral bleaching observed in sector 3.
Dugong feeding observed near seagrass bed.
Hawksbill turtle nesting activity on north beach.
Follow up required for bleaching assessment.
"""

print("Field Notes Analysis")
print("=" * 40)

# Word count
words = field_notes.split()
print(f"Total words: {len(words)}")

# Line count
lines = field_notes.strip().split("\n")
print(f"Total lines: {len(lines)}")

# Search for key terms
keywords = ["turtle", "bleaching", "coral",
            "temperature", "dugong"]

print("\nKeyword Search:")
for keyword in keywords:
    count = field_notes.lower().count(keyword)
    found = "✅ Found" if count> 0 else "❌ Not found"
    print(f" {keyword:<15} {found} ({count}x)")

# Extract observer name
for line in lines:
    if "Observer:" in line:
        observer = line.replace("Observer:","").strip()
        print(f"\nObserver: {observer}")
    if "Location:" in line:
        location = line.replace("Location:","").strip()
        print(f"\nLocation: {location}")
```
