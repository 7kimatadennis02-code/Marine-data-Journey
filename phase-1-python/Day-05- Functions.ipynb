## Function with Calculation

```python
# Convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

# Convert Fahrenheit to celsius
def fahrenheit_to_celsius(fahrenheit):
    celsius = (fahrenheit -32) * 5/9
    return celsius

# Use the functions
temp1 = celsius_to_fahrenheit(27.5)
temp2 = celsius_to_fahrenheit(31.0)
temp3 = fahrenheit_to_celsius(98.6)

print(f"27.5°C = {temp1:.1f}°F")
print(f"31.0°C = {temp2:.1f}°F")
print(f"98.6°F = {temp3:.1f}°C")
```

## Function with multiple parameters

```python
# Calculate fish population density
def population_density(fish_count, area_km2):
    density = fish_count / area_km2
    return density

# Calculate percentage
def calculate_percentage(part, total):
    percentage = (part / total) * 100
    return percentage

#Use the functions
density1 = population_density(350, 25)
density2 = population_density(120, 8)

coverage = calculate_percentage(142, 536)

print(f"site 1 density: {density1:.1f} fish/km2")
print(f"Site 2 density: {density2:.1f} fish/km2")
print(f"Protected coastline: {coverage:.1f}%")
```

## Function with conditional logic

```python
# Classify water temperature
def classify_temperature(temp):
    if temp > 30:
        return "Too high - bleaching risk"
    elif temp < 20:
        return "Too low - stress risk"
    else:
        return "Normal range"

# Classify pH
def classify_ph(ph):
    if ph < 6.5:
        return "Acidic - danger"
    elif ph > 8.5:
        return "Alkalne - danger"
    else:
        return "Normal"

# Use the functions
temperatures = [26.5, 31.2, 28.0, 19.8, 32.1]
ph_levels = [7.2, 6.1, 8.9, 7.8, 8.3]

print("Temperature Classifications:")
for temp in temperatures:
    result = classify_temperature(temp)
    print(f" {temp}°C - {result}")

print("\npH Classifications:")
for ph in ph_levels:
    result = classify_ph(ph)
    print(f" pH {ph} - {result}")
```

## Function with Default Parameters

```python
# Default parameter means the value is used
# if no argument is provided
def assess_fish_population(count, threshold = 50):
    if count >= threshold:
        return f"{count} fish - population healthy"
    else:
        return f"{count} fish - Below threshold of {threshold}"

# Using default threshold
print(assess_fish_population(75))
print(assess_fish_population(32))

# Using custom threshold
print(assess_fish_population(75, threshold=100))
print(assess_fish_population(32, threshold=25))
```

## Function Returning Multiple Values

```python
# Return multiple values at once
def summarize_readings(readings):
    total = sum(readings)
    average = total/len(readings)
    minimum = min(readings)
    maximum = max(readings)
    return total, average, minimum, maximum

# Use the function
catch_data = [120, 95, 140, 88, 110, 155, 72]

total, avg, low, high =summarize_readings(catch_data)

print(f"Total catch:{total} fish")
print(f"Average catch: {avg:.1f} fish")
print(f"Lowest day: {low} fish")
print(f"Highest day: {high} fish")
```

## Practice - Marine Analysis Toolkit

```python
# Build a set of reusable marine analyst functions
def calculate_average(readings):
    return sum(readings) / len(readings)

def flag_outliers(readings, threshold):
    outliers = []
    for value in readings:
        if value > threshold:
            outliers.append(value)
    return outliers

def site_report(site_name, temperatures, ph_readings):
    avg_temp = calculate_average(temperatures)
    avg_ph = calculate_average(ph_readings)
    temp_status = classify_temperature(avg_temp)
    ph_status = classify_ph(avg_ph)

    print(f"\nSite Report: {site_name}")
    print("=" * 35)
    print(f"Average temperature; {avg_temp:.1f}°C - {temp_status}")
    print(f"Average pH: {avg_ph:.2f} - {ph_status}")
    print(f"Temperature outliers above 30°C: "
          f"{flag_outliers(temperatures, 30)}")

# Run reports for two sites
site_report(
    "Watamu Marine Reserve",
    [27.5, 28.1, 31.2, 26.8, 29.4],
    [7.8, 8.1, 7.6 ,7.9 ,8.0]
)

site_report(
    "Malindi Marine Reserve",
    [29.1, 32.4, 30.8, 31.5, 28.9],
    [6.3, 7.1, 6.8, 7.4, 6.9]
)
```
