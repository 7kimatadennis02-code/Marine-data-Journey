## Basic for Loop

```python
# Loop through a list of marine sites
sites = ["Watamu", "Mombasa", "Malindi", "Kilifi", "Lamu"]

for site in sites:
    print(F"Monitoring site: {site}")
    
```

# for Loop with range()

```python
# range(start, stop, step)
# Count montoring days
print("Monitoring schedule:")
for day in range(1, 8):
    print(f"Day {day} -field survey")

print("---")

# Count in steps
print("Water sampling every 3 days:")
for day in range(1, 31, 3):
    print(f"Sample on day {day}")
```

## Loop with Conditionals Inside

```python
# Check each water temperature reading
temperatures = [26.5, 31.2, 28.0, 30.5, 24.8, 32.1, 27.3]

print("Temperature Assessment:")
print("=" * 30)

for temp in temperatures:
    if temp > 30:
        print(f"{temp}°C - Warning: Bleaching risk")
    elif temp < 25:
        print(f"{temp}°C - Caution: Below normal")
    else:
        print(f"{temp}°C - Normal")
```

## Counting and Accumulating in Loops

```python
# Fish catch over a week
daily_catches = [120, 95, 140,88,110,155,72]

total = 0
high_catch_days = 0
low_catch_days = 0

for catch in daily_catches:
    total = total + catch
    if catch > 120:
        high_catch_days = high_catch_days + 1
    elif catch < 90:
        low_catch_days = low_catch_days + 1

average = total / len(daily_catches)

print(f"Total catch: {total} fish")
print(f"Average daily catch: {average:.1f} fish")
print(f"High catch days: {high_catch_days}")
print(f"Low catch days: {low_catch_days}")
```

## While Loop

```python
# Simulate water level rising
water_level = 0
max_level = 10

print("Monitoring water level rise:")
while water_level < max_level:
    water_level =water_level + 1
    print(f"Water level: {water_level}m")

print("Maximum level reached - alert triggered")
```

## break and continue

```python
# break - stop the loop early
print("Searching for critical reading:")
readings = [7.2, 7.8, 6.1, 8.9, 7.5]

for reading in readings:
    if reading < 6.5 or reading > 8.5:
        print(f"Critical reading found: {reading}")
        print ("Stopping search")
        break
    print(f"{reading} - Normal")

# continue - skip this item and move to the next
print("Skipping low readings")
catches = [120, 45, 98, 30, 140, 85, 200]

for catch in catches:
    if catch < 50:
        print(f"{catch} - Too low, skipping")
        continue
    print(f"{catch} fish - processing")
      
```
