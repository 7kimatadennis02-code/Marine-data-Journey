## Exercse 1 _ Fish Catch Summary

```python
# weekly fish catch data
monday = 120
tuesday = 95
wednesday = 140
thursday = 88
friday = 110
saturday = 155
sunday = 72

# calculations
total_catch = monday + tuesday + wednesday + thursday + friday + saturday + sunday
average_catch = total_catch / 7
best_day = saturday
worst_day = sunday

print(f"Total weekly catch: {total_catch} fish")
print(f"Average daily catch: {average_catch:.1f} fish")
print(f"Best single catch day: {best_day} fish")
print(f"Worst single catch day: {worst_day} fish")
print(f"Above average today (Friday): {friday > average_catch}")
```

## Exercise 2 - Water quality Assessment

```python
# Water quality readings at Watamu Marine Reserve
site = "Watamu Marine Reserve"
ph = 7.8
dissolved_oxygen = 6.5
turbidity = 12.3
water_temp = 28.2
salinity = 35.1

# Safe thresholds
safe_ph_min = 6.5
safe_ph_max = 8.5
safe_oxygen_min = 5.0
safe_temp_max = 30.0

# Assessments
ph_safe = safe_ph_min <= ph <= safe_ph_max
oxygen_safe = dissolved_oxygen >= safe_oxygen_min
temp_safe =water_temp <= safe_temp_max

print(f"site: {site}")
print(f"pH level: {ph} - Safe: {ph_safe}")
print(f"Dssolved oxygen: {dissolved_oxygen} mg/L - Safe: {oxygen_safe}")
print(f"Turbidity: {turbidity} NTU")
print(f"Water tempertature: {water_temp}°C - Safe: {temp_safe}")
print(f"Salinity: {salinity} ppt")
```

## Exercise 3 - Marine Protected Area Coverage

```python
# Kenya coastaline data
total_coastline_km = 536
protected_coastline_km = 142
unprotected_coastline_km = total_coastline_km - protected_coastline_km

# Percentage calculations
protected_percentage = (protected_coastline_km / total_coastline_km) * 100
unprotected_percentage = (unprotected_coastline_km / total_coastline_km) * 100

# Target
conservation_target = 30.0
target_met = protected_percentage >= conservation_target

print(f"Total Kenya coastline: {total_coastline_km} km")
print(f"Protected coastline: {protected_coastline_km} km")
print(f"Unprotected coastline: {unprotected_coastline_km} km")
print(f"Protection coverage: {protected_percentage:.1f}% ")
print(f"unprotected coverage: {unprotected_percentage:.1f}% ")
print(f"30% conservation met: {target_met}")

```

## Exercise 4 -Species Population Estimate

```python
# Green turtle population monitoring
species = "Green Turtle"
survey_area_km2 = 25
turtles_observed = 38
survey_duration_days = 14

# Estimates
density_per_km2 = turtles_observed // survey_area_km2
daily_sightings = turtles_observed // survey_duration_days

# Conservation status
minimum_viable_population = 50
population_concern = turtles_observed < minimum_viable_population

print(f"Species: {species}")
print(f"Survey area: {survey_area_km2} km2")
print(f"Turtles Observed: {turtles_observed} turtles")
print(f"Survey duration: {survey_duration_days} days")
print(f"Population density: {density_per_km2} turtles/km2")
print(f"Average daily sightings: {daily_sightings} turtles")
print(f"Population concern flagged: {population_concern}")
```
