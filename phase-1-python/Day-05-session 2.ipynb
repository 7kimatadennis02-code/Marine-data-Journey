## Exercise 1 - Unit conversion toolkit

```python
# A collection of useful unit conversion functions

def knots_to_kmh(knots):
    return knots * 1.852

def kmh_to_knots(kmh):
    return kmh / 1.852

def meters_to_feet(meters):
    return meters * 3.28084

def feet_to_meters(feet):
    return feet / 3.28084

def kg_to_pounds(kg):
    return kg * 2.20462

def pounds_to_kg(pounds):
    return pounds / 2.20462

def nautical_miles_to_km(nm):
    return nm * 1.852

# Test all conversions
print("Unit Conversion Toolkit")
print("=" * 35)
print(f"15 knots = {knots_to_kmh(15):.1f} km/h")
print(f"30km/h = {kmh_to_knots(30):.1f} knots")
print(f"100 meters = {meters_to_feet(100):.1f} feet")
print(f"50 feet = {feet_to_meters(50):.1f} meters")
print(f"25 kg fish = {kg_to_pounds(25):.1f} pounds")
print(f"10 nauticak miles = {nautical_miles_to_km(10):.1f} km")
```

## Exercise 2 - Water Quality Scorer

```python
# Score each parameter out of 10
# then combine into overall score

def score_ph(ph):
    if 7.0 <= ph <= 8.0:
        return 10
    elif 6.5 <= ph < 7.0 or 8.0 < ph <= 8.5:
        return 6
    else:
        return 2

def score_temperature(temp):
    if 24 <= temp <= 28:
        return 10
    elif 20 <= temp < 24 or 28 < temp <= 30:
        return 6
    else:
        return 2

def score_oxygen(oxygen):
    if oxygen >= 7.0:
        return 10
    elif oxygen >= 5.0:
        return 6
    else:
        return 2

def overall_water_quality(site, ph, temp, oxygen):
    ph_score = score_ph(ph)
    temp_score = score_temperature(temp)
    oxygen_score = score_oxygen(oxygen)

    total_score = (ph_score + temp_score + oxygen_score) / 3

    if total_score >= 8:
        grade = "Exellent"
    elif total_score >= 6:
        grade = "Good"
    elif total_score >= 4:
        grade = "fair"
    else:
        grade = "poor"

    print(f"\nWater Quality Report : {site}")
    print("=" * 35)
    print(f"pH score: {ph_score}/10")
    print(f"Temperature score: {temp_score}/10")
    print(f"Oxygen score: {oxygen_score}/10")
    print(f"Overall score: {total_score:.1f}/10 - {grade}")

# Run for three sites
overall_water_quality("Watamu", 7.8, 27.5, 6.2)
overall_water_quality("Malindi", 6.3, 31.2, 4.8)
overall_water_quality("Kilifi", 7.5, 26.8,7.1)
```

## Exercise 3 - catch statistics function

```python
# A complete statistiics function for any datasets

def catch_statistics(site, catches):
    total = sum(catches)
    average = total / len(catches)
    minimum = min(catches)
    maximum = max(catches)
    spread = maximum - minimum

    # Count good and poor days
    good_days = 0
    poor_days = 0
    for catch in catches:
        if catch >= average:
            good_days +=1
        else:
            poor_days +=1

    print(f"\nCatch Statistics: {site}")
    print("=" * 35)
    print(f"Total catch: {total} fish")
    print(f"Average: {average:.1f} fish/day")
    print(f"Best day: {maximum} fish")
    print(f"Worst day: {minimum} fish")
    print(f"Spread: {spread} fish")
    print(f"Above average: {good_days} days")
    print(f"Below average: {poor_days} days")

# Run for two fish sites
catch_statistics(
    "Malindi Fishing Zone",
        [142, 98, 175, 63, 210, 88, 156]
)

catch_statistics(
    "Kilifi Fishing Zone",
        [88, 112, 95, 134, 78, 145, 103]
)
            
    
```

## Exercise 4  - conservation status checker

```python
# Check conservation status of multple species

def conservation_status(species, population, 
                        trend, habitat_loss_percent):

    # Score the population
    if population >= 10000:
        pop_score = "Stable"
    elif population > 1000:
        pop_score = "Vulnerable"
    elif population >= 100:
        pop_score = "Endangered"
    else:
        pop_score = "Critically Endangered"

    # Assess trend
    if trend == "increasing":
        trend_note = "population growing"
    elif trend == "stable":
        trend_note = "population holding"
    else:
        trend_note = "population declining"

    # Habitat concern
    if habitat_loss_percent >= 50:
        habitat_note = "Severe habitat loss"
    elif habitat_loss_percent >= 25:
        habitat_note = "Moderate habitat loss"
    else:
        habitat_note = "Habitat relatively intact"

    print(f"\nSpecies: {species}")
    print("=" * 35)
    print(f"population: {population} - {pop_score}")
    print(f"Trend: {trend_note}")
    print(f"Habitat: {habitat_note}")

# Check three species
conservation_status("Green Turtle", 4500,
                    "increasing", 18)

conservation_status("Dudong", 680,
                    "declining", 42)

conservation_status("Hawksbill Turtle", 85,
                    "declining", 61)
        
```

## Exercise 5 - Tidal Zone Classifier

```python
# Classify marine zones based on depth

def classify_depth_zone(depth_m):
    if depth_m <= 0:
        return "Supratidal - above high tide"
    elif depth_m <= 5:
        return "Intertidal - between tides"
    elif depth_m <=30:
        return "Subtidal shallow - sunlit zone"
    else: 
        return "Deep sea - no sunlight"

def survey_depth_zones(site, depth_readings):
    print(f"\nDepth Zone Survey: {site}")
    print("=" * 40)

    zone_counts = {}

    for depth in depth_readings:
        zone = classify_depth_zone(depth)
        print(f" {depth:>6}m - {zone}")

        if zone in zone_counts:
            zone_counts[zone] += 1
        else:
            zone_counts[zone] = 1

    print("---")
    print("Zone Summary:")
    for zone, count in zone_counts.items():
        print(f" {zone}: {count} readings")

# Survey two locations
survey_depth_zones(
    "Watamu Reef Survey",
    [0.5, 3.2, 12.0, 28.5, 45.0, 8.3, 2.1,180.0]
)

survey_depth_zones(
    "Malindi Transact",
    [1.0, 6.5, 15.0, 95.0, 4.2, 22.0, 350.0]
)
```
