## Marine species Database

## Exercise 2 - Survey Comparison Tool

```python
# Compare species across three survey zones
zone_north = {"Dugong", "Green Turtle", "Dolphin",
              "Manta Ray", "Whale Shark"}

zone_central = {"Green Turtle", "Dolphin",
                "Hawksbill Turtle", "Reef Shark"}

zone_south = {"Dugong", "Humpback Whale",
              "Green Turtle", "Manta Ray"}

# Analysis
all_species = zone_north | zone_central | zone_south
widespread = zone_north & zone_central & zone_south
only_north = zone_north - zone_central - zone_south
only_central = zone_central - zone_north - zone_south
only_south = zone_south - zone_north - zone_central

print("Survey Zone Comparison")
print("=" * 40)
print(f"Total unique species: {len(all_species)}")
print(f"All species: {all_species}")
print(f"\nFound in ALL zones: {widespread}")
print(f"\nOnly in North: {only_north}")
print(f"Only in Central: {only_central}")
print(f"Only in South: {only_south}")

# Count species per zone
print("\nSpecies count per zone:")
print(f" North zone:   {len(zone_north)} species")
print(f" Central zone: {len(zone_central)} species")
print(f" South zone:   {len(zone_south)} species")
```

## Exercise 3 - Monitoring station Registry

```python
# Registry of monitoring stations
stations = {
    "WTM001": {
        "name": "Watamu North Buoy",
        "coords": (-3.2853, 40.0312),
        "active": True,
        "readings": [27.5,28.1,27.8,29.2,28.5]

    },
    "MLN002": {
        "name": "Malindi Reef Station",
        "coords": (-3.1938, 40.1420),
        "active": True,
        "readings": [31.2, 30.8, 32.1, 29.8, 31.5]

    },
    "KLF003": {
        "name": "Kilifi Channel Buoy",
        "coords": (-3.6297, 39.8499),
        "active": False,
        "readings": [26.8,27.1,26.5,27.4,26.9]

    },
    "MBS004": {
        "name": "Mombasa Harbour Monitor",
        "coords": (-4.0512, 39.6720),
        "active": True,
        "readings": [29.4,30.1,28.8, 29.7,30.3]
    },
}

print("Monitoring Station Registry")
print("=" * 45)

for code, station in stations.items():
    name = station["name"]
    active = "🟢 Active" if station["active"] else "🔴 offline"
    readings = station["readings"]
    avg_temp = sum(readings) / len(readings)

    print(f"\n{code} - {name}")
    print(f" status: {active}")
    print(f" Coords: {station['coords']}")
    print(f" Avg temp: {avg_temp:.1f}°C")

    if avg_temp > 30:
        print(f" ⚠️ High temperature alert")
```

## Exercise 4 - Catch Log Analyzer

```python
# Weekly catch log - list of tuples
# Each tuple is (day, species, weight_kg, location)
catch_log = [
    ("Monday", "Yellowfin Tuna", 45.2, "Malindi"),
    ("Monday", "Sailfish", 32.8, "Watamu"),
    ("Tuesday", "Yellowfin Tuna", 38.5, "Malindi"),
    ("Tuesday", "Kingfish", 22.1, "Kilifi"),
    ("Wednesday", "Sailfish", 41.0, "Watamu"),
    ("Wednesday", "Yellowfin Tuna", 29.8, "Malindi"),
    ("Thursday", "Kingfish", 18.4, "Kilifi"),
    ("Thursday", "Yellowfin Tuna", 52.3, "Malindi"),
    ("Friday", "Sailfish", 37.6, "Watamu"),
    ("Friday", "Kingfish", 25.9, "Kilifi"),
]

# Analyze by species
species_totals = {}
location_totals = {}

for day, species,weight,location in catch_log:
    # Accumulate by species
    if species in species_totals:
        species_totals[species] += weight
    else:
        species_totals[species] = weight

    # Accumulate by location
    if location in location_totals:
        location_totals[location] += weight
    else:
        location_totals[location] = weight

print("Weekly Catch Analysis")
print("=" * 35)
print("\nBy Species:")
for species, total in species_totals.items():
    print(f" {species:<20} {total:.1f} kg")

print("\nBy Location:")
for location, total in location_totals.items():
    print(f" {location:<15} {total:.1f} kg")

print(f"\nTotal Catch: "
      f"{sum(species_totals.values()):.1f} kg")
```
