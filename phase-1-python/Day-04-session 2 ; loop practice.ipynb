## Exercise 1 - Rainfall Pattern Printer

```python
# Monthly rainfall data for Kenya coast (mm)
months =["Jan", "Feb", "Mar", "Apr", "May", "Jun",
         "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

rainfall = [51, 33, 64, 196, 320, 87,
            61, 74, 65, 109, 97, 58]

print("Kenya Coast - Monthly Rainfall Report")
print("=" * 40)

for i in range(len(months)):
    month = months[i]
    rain = rainfall[i]
    bar = "█" * (rain // 20)

    if rain >= 200:
        status = "🌧 Heavy"
    elif rain >= 100:
        status = "🌦 Moderate"
    elif rain >= 50:
        status = "🌤 Light"
    else:
        status = "☀️ Dry"

    print(f"{month}: {rain:>4} mm {bar} {status}")
```

## Exercise 2 - Species survey processor

```python
# Species sightings at Watamu Marine Reserve
species_data = [
    ["Green Turtle", 12],
    ["Hawksbill Turtle", 4],
    ["Dolphin", 28],
    ["Dugong", 1],
    ["Whale shark", 3],
    ["Manta Ray", 9],
    ["Humpback whale", 2],
]

print("Species Sighting Report - Watamu Marine Reserve")
print("=" * 48)

total_sightings = 0
rare_species = []
common_species =[]

for species,count in species_data:
    total_sightings = total_sightings + count

    if count <= 3:
        status = "RARE - priority monitoring"
        rare_species.append(species)
    elif count <= 10:
        status = "Occasional"
        common_species.append(species)
    else:
        status = "Frequent"
        common_species.append(species)

    print(f"{species:<20} {count:>3} sightings _ {status}")

print("=" * 48)
print(f" Total sightings: {total_sightings}")
print(f"Rare species count: {len(rare_species)}")
print(f"Rare species: {rare_species}")
```

## Exercise 3 - Water quality multi site checker

```python
# Multiple site readings
sites_data = [
    ["Watamu", 7.8, 27.5, 6.2],
    ["Malindi", 6.3,29.1,4.8],
    ["Mombasa", 8.1, 31.2,5.5],
    ["Kilifi", 7.5,26.8,7.1],
    ["Lamu", 8.7, 28.4, 4.2],
]

# site, ph, temperature, dissolved oxygen

print("Multi-Site Water Quality Assessment")
print("=" *45)

alert_sites = []

for site, ph, temp, oxygen in sites_data:
    issues = []

    if ph < 6.5 or ph > 8.5:
        issues.append(f"pH {ph} out of range")
    if temp > 30:
        issues.append(f"Temp {temp}°c too high")
    if oxygen < 5.0:
        issues.append(f"Oxygen {oxygen} mg/L too low")

    if issues:
        status = "⚠️ Alert"
        alert_sites.append(site)
    else:
        status = "✅ Normal"

    print(f"{site:<10} {status}")
    for issue in issues:
        print(f"         › {issue}")

print("=" *45)
print(f"Sites requiring attention: {len(alert_sites)}")
print(f"Alert sites: {alert_sites}")
```

## Exercise 4 - Cumulative catch tracker

```python
# Daily fish catch over two weeks
daily_catch = [142, 98, 175, 63, 210, 88, 156,
               134, 77, 195, 112, 48, 188, 101]

print("Two Week Cumulative Catch Tracker")
print("=" * 45)
print(f"{'Day':<6} {'Catch':>7} {'Cumulative':>12} {'Status':<20}")
print("=" * 45)

cumulative = 0
above_target = 0 
below_target = 0
target = 120

for day, catch in enumerate(daily_catch, start=1):
    cumulative = cumulative + catch

    if catch >= target:
        status = "✅ On target"
        above_target += 1
    else:
        status = "❌ Below target"
        below_target += 1

    print(f"{day:<6} {catch:>7} {cumulative:>12} {status}")

print("=" *45)
print(f"Average daily catch: {cumulative/len(daily_catch):.1f} fish")
print(f"Days on target: {above_target}")
print(f"Days below target: {below_target}")
print(f"Total catch: {cumulative} fish")
```

## Exercise 5 - Nesting season counter

```python
# Turtle nesting recirds for the seaon
# Each entry is [beach, nests_found, eggs_per_nest]
nesting_records = [
    ["Watamu Beach", 8, 112],
    ["Malindi Beach", 3, 98],
    ["Lamu Beach", 12, 125],
    ["Kilifi Beach", 5, 108],
    ["Mombasa Beach", 2, 95],
    ["Shimba Beach", 7, 118],
]

print("Sea Turtle Nesting Season Summary")
print("=" * 45)

total_nests = 0
total_eggs = 0
best_beach = ""
best_count = 0

for beach, nests, eggs_per_nest in nesting_records:
    estimated_eggs = nests * eggs_per_nest
    total_nests += nests
    total_eggs += estimated_eggs

    if nests > best_count:
        best_count = nests
        best_beach = beach

    print(f"{beach:<20} {nests:>3} nests -"
          f"~{estimated_eggs:>5} eggs estimated")

print("=" * 45)
print(f"Total nests recorded: {total_nests}")
print(f"Total eggs estimated: {total_eggs}")
print(f"Best nesting beach: {best_beach} ({best_count} nests)")
         
```
