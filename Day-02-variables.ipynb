## Data Types

```python
# Integer - whole numbers
fish_count = 142
water_depth = 30

# Float - decimal numbers
water_temp = 27.5
salinity = 34.8
ph_level = 8.1

# String - text
site_name = "Mombasa Marine Researve"
species = "Green Turtle"
researcher = "Dennis Kimata"

# Boolean - true or False
sample_collected = True
endangered = False

# Print labels and units
print("Fish count:", fish_count, "individuals")
print("Water temperature:", water_temp, "°C")
print("Water depth:", water_depth, "m")
print("Salinity:", salinity, "ppt")
print("pH level:", ph_level)
print("Site:", site_name)
print("Species:", species)
print("Researcher:", researcher)
print("Sample collected:", sample_collected)
print("Endangered species:", endangered) 
```

##Checking Data Types

```python
print(type(fish_count))
print(type(water_temp))
print(type(site_name))
print(type(sample_collected))
```

```python
## Arithmetc Operators
```

```python
# Additon
total_fish = 142 + 98
print("Total fish:", total_fish)

# Subtraction
temp_difference = 29.5 - 27.5
print("Temperature difference:", temp_difference)

# Multiplication
area_hectares =4.5 *10
print("Area in hectares:", area_hectares)

# Division
average_catch = 450 / 7
print("Average daily catch:", average_catch)

# Floor division - whole number result
full_days = 450 // 7
print("Full days:", full_days)

# Modulo - remander
remaining = 450 % 7
print("Remaining fish:", remaining)

#Exponent - power
square_area = 5 ** 2
print("Square are:", square_area)
```

## Comparison Operators

```python
water_temp = 27.5
safe_temp = 30

print(water_temp > safe_temp)   # Greater than
print(water_temp < safe_temp)   # Less than
print(water_temp == safe_temp)  # Equal to
print(water_temp != safe_temp)  # Not equal to
print(water_temp >= 27.5)       # Greater than or equal
print(water_temp <= 27.5)       # Less than or equal
```

## Practice - Marine site calculations

```python
# Site data
site_name = "Watamu Marine Reserve"
total_rainfall_mm = 1250
monitoring_days = 30
fish_species_count = 87
water_temp_celsius = 27.5

#Site name
site_name = "Watamu Marine Researve"
print("Site:", site_name)
#Calculations
daily_rainfall = total_rainfall_mm / monitoring_days
print("Daily average rainfall:", round(daily_rainfall, 2), "mm")
temp_farenheit = (water_temp_celsius * 9/5) + 32
print("Water temperature:", temp_farenheit, "°F")
is_diverse = fish_species_count > 50
print("High biodiversity site:", is_diverse)
```

## Calculation using f-strings

```python
print(f"Site: {site_name}")
daily_rainfall = total_rainfall_mm / monitoring_days
print(f"Daily average rainfall: {daily_rainfall:.2f} mm")
temp_farenheit = (water_temp_celsius * 9/5) + 32
print(f"Water temperature: {temp_farenheit} °F")
is_diverse = fish_species_count > 50
print(f"High biodiversity site: {is_diverse}")
```
